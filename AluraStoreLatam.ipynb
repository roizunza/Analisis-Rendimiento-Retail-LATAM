### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# calcular la facturación total (suma de la columna 'Precio') para cada tienda
facturacion_t1 = tienda['Precio'].sum()
facturacion_t2 = tienda2['Precio'].sum()
facturacion_t3 = tienda3['Precio'].sum()
facturacion_t4 = tienda4['Precio'].sum()
```

```python
# imprimir los resultados para comparar
print(f"Facturación Tienda 1: ${facturacion_t1:,.2f}")
print(f"Facturación Tienda 2: ${facturacion_t2:,.2f}")
print(f"Facturación Tienda 3: ${facturacion_t3:,.2f}")
print(f"Facturación Tienda 4: ${facturacion_t4:,.2f}")
```

# 2. Ventas por categoría

```python
import matplotlib.pyplot as plt

# 1. Función simple para analizar y graficar cada tienda
def analizar_categorias(df, nombre_tienda):
    print(f"\n--- {nombre_tienda} ---")
    conteo = df['Categoría del Producto'].value_counts()
    print(conteo.head(3)) # Muestra solo las top 3 categorías en texto

    # Visualización básica (Capa visual)
    plt.figure(figsize=(10, 4))
    conteo.plot(kind='bar', title=f'Ventas por Categoría - {nombre_tienda}')
    plt.ylabel('Cantidad de Ventas')
    plt.xticks(rotation=45)
    plt.show()

# 2. Ejecutar el análisis para las 4 tiendas
analizar_categorias(tienda, "Tienda 1")
analizar_categorias(tienda2, "Tienda 2")
analizar_categorias(tienda3, "Tienda 3")
analizar_categorias(tienda4, "Tienda 4")
```

# 3. Calificación promedio de la tienda


```python
import matplotlib.pyplot as plt

# 1. Cálculo de promedios
promedio_t1 = tienda['Calificación'].mean()
promedio_t2 = tienda2['Calificación'].mean()
promedio_t3 = tienda3['Calificación'].mean()
promedio_t4 = tienda4['Calificación'].mean()

# 2. Impresión de datos
print(f"Promedio Tienda 1: {promedio_t1:.2f} estrellas")
print(f"Promedio Tienda 2: {promedio_t2:.2f} estrellas")
print(f"Promedio Tienda 3: {promedio_t3:.2f} estrellas")
print(f"Promedio Tienda 4: {promedio_t4:.2f} estrellas")

# 3. Visualización comparativa
promedios = [promedio_t1, promedio_t2, promedio_t3, promedio_t4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
colores = ['#4CAF50', '#FF9800', '#2196F3', '#F44336'] # Colores diferente por tienda

plt.figure(figsize=(8, 5))
plt.bar(tiendas, promedios, color=colores)
plt.title('Satisfacción Promedio del Cliente por Tienda')
plt.ylabel('Calificación (1-5)')
plt.ylim(0, 5)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Añadir el valor encima de cada barra
for i, v in enumerate(promedios):
    plt.text(i, v + 0.1, f"{v:.2f}", ha='center', fontweight='bold')

plt.show()
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt



def analizar_best_sellers(df, nombre_tienda):

    # Contamos la frecuencia de cada producto

    conteo_productos = df['Producto'].value_counts()



    # Sepaación de los mejores (Top 5) y los peores (Bottom 5)

    top_5 = conteo_productos.head(5)

    bottom_5 = conteo_productos.tail(5)



    plt.figure(figsize=(10, 5))

    # Inversión del orden (.iloc[::-1]) para que el #1 quede arriba en el gráfico horizontal

    top_5.iloc[::-1].plot(kind='barh', color='skyblue')

    plt.title(f'Top 5 Productos Más Vendidos - {nombre_tienda}')

    plt.xlabel('Cantidad de Unidades Vendidas')

    plt.grid(axis='x', linestyle='--', alpha=0.5)

    plt.show()


    print(f"Productos MENOS vendidos en {nombre_tienda}:")

    # índice (nombre) y valor (cantidad)

    for producto, cantidad in bottom_5.items():

        print(f"   - {producto}: {cantidad} ventas")

    print("-" * 50)



# Análisis para las 4 tiendas

analizar_best_sellers(tienda, "Tienda 1")

analizar_best_sellers(tienda2, "Tienda 2")

analizar_best_sellers(tienda3, "Tienda 3")

analizar_best_sellers(tienda4, "Tienda 4")
```

# 5. Envío promedio por tienda

```python
import matplotlib.pyplot as plt

def analizar_dispersión_ventas(df, nombre_tienda):
    resumen = df.groupby('Producto').agg({
        'Precio': 'mean',
        'Calificación': 'count'
    }).reset_index()

    resumen.columns = ['Producto', 'Precio_Promedio', 'Cantidad_Ventas']
    resumen = resumen.sort_values(by='Cantidad_Ventas', ascending=False)

    # Configuración del Gráfico de Dispersión
    plt.figure(figsize=(10, 6))

    # Eje X = Precio, Eje Y = Cantidad
    plt.scatter(resumen['Precio_Promedio'], resumen['Cantidad_Ventas'],
                alpha=0.6, c='blue', edgecolors='black', s=80)

    plt.title(f'Dispersión: Precio vs. Demanda - {nombre_tienda}')
    plt.xlabel('Precio del Producto ($)')
    plt.ylabel('Cantidad de Unidades Vendidas')
    plt.grid(True, linestyle='--', alpha=0.3)

    # Top 3 más vendidos (Color Verde)
    for i in range(3):
        row = resumen.iloc[i]
        plt.annotate(row['Producto'],
                     (row['Precio_Promedio'], row['Cantidad_Ventas']),
                     xytext=(5, 5), textcoords='offset points',
                     fontsize=9, color='green', fontweight='bold')

    # Bottom 1 menos vendido (Color Rojo)
    peor_producto = resumen.iloc[-1]
    plt.annotate(f"Menos: {peor_producto['Producto']}",
                 (peor_producto['Precio_Promedio'], peor_producto['Cantidad_Ventas']),
                 xytext=(5, -10), textcoords='offset points',
                 fontsize=8, color='red')

    plt.show()

    print(f"--- Datos Clave {nombre_tienda} ---")
    print(f"Más vendido: {resumen.iloc[0]['Producto']} ({resumen.iloc[0]['Cantidad_Ventas']} u.)")
    print(f"Menos vendido: {resumen.iloc[-1]['Producto']} ({resumen.iloc[-1]['Cantidad_Ventas']} u.)")
    print("-" * 50)

analizar_dispersión_ventas(tienda, "Tienda 1")
analizar_dispersión_ventas(tienda2, "Tienda 2")
analizar_dispersión_ventas(tienda3, "Tienda 3")
analizar_dispersión_ventas(tienda4, "Tienda 4")
```

6. Reporte final

```python
import matplotlib.pyplot as plt
import pandas as pd

data_tiendas = [
    {'nombre': 'Tienda 1', 'df': tienda},
    {'nombre': 'Tienda 2', 'df': tienda2},
    {'nombre': 'Tienda 3', 'df': tienda3},
    {'nombre': 'Tienda 4', 'df': tienda4}
]

# Métricas clave para cada tienda
resumen = []
for t in data_tiendas:
    df = t['df']
    resumen.append({
        'Tienda': t['nombre'],
        'Ingresos': df['Precio'].sum(),
        'Calificacion_Prom': df['Calificación'].mean(),
        'Envio_Prom': df['Costo de envío'].mean(),
        'Ventas_Totales': len(df),
        'Dataframe': df
    })

df_resumen = pd.DataFrame(resumen).set_index('Tienda')

# tienda con menor rendimiento (Ingresos)
peor_tienda_nombre = df_resumen['Ingresos'].idxmin()
peor_tienda_data = df_resumen.loc[peor_tienda_nombre]
df_peor_tienda = data_tiendas[int(peor_tienda_nombre.split()[-1])-1]['df']


# GENERACIÓN DE GRÁFICOS
fig, axes = plt.subplots(1, 3, figsize=(20, 6))
plt.suptitle(f'Análisis de Rendimiento: Foco en {peor_tienda_nombre}', fontsize=16)

# GRÁFICO 1: BARRAS-Comparación de Ingresos
colores = ['grey' if x != peor_tienda_nombre else 'red' for x in df_resumen.index]
axes[0].bar(df_resumen.index, df_resumen['Ingresos'], color=colores)
axes[0].set_title('Ingresos Totales por Tienda')
axes[0].set_ylabel('Monto ($)')
axes[0].grid(axis='y', linestyle='--', alpha=0.3)

# GRÁFICO 2: PASTEL-Categorías de la peor tienda
cat_counts = df_peor_tienda['Categoría del Producto'].value_counts()
axes[1].pie(cat_counts, labels=cat_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)
axes[1].set_title(f'Mix de Categorías ({peor_tienda_nombre})')

# GRÁFICO 3: DISPERSIÓN -Productos de la PEOR tienda
# Precio vs Cantidad para ver si tiene productos estrella
prod_stats = df_peor_tienda.groupby('Producto').agg({'Precio': 'mean', 'Calificación': 'count'}).reset_index()
prod_stats.columns = ['Producto', 'Precio_Prom', 'Ventas']
axes[2].scatter(prod_stats['Precio_Prom'], prod_stats['Ventas'], color='red', alpha=0.6, edgecolors='black')
axes[2].set_title(f'Dispersión: Precio vs Ventas ({peor_tienda_nombre})')
axes[2].set_xlabel('Precio Promedio')
axes[2].set_ylabel('Cantidad Vendida')
axes[2].grid(True, linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()

# 3. REDACCIÓN DEL INFORME (SALIDA DE TEXTO)
print("\n" + "="*80)
print(f"INFORME EJECUTIVO: RECOMENDACIÓN DE VENTA DE ACTIVO")
print("="*80)

print(f"\n1. INTRODUCCIÓN")
print("El objetivo de este análisis fue evaluar el desempeño financiero y operativo de las 4 tiendas")
print("pertenecientes al grupo, con el fin de identificar el activo con menor rendimiento para su desinversión.")

print(f"\n2. HALLAZGOS PRINCIPALES")
print(f"- Análisis Financiero: La {peor_tienda_nombre} presenta la facturación más baja del grupo,")
print(f"  con un total de ${peor_tienda_data['Ingresos']:,.2f}.")
print(f"- Satisfacción del Cliente: Su calificación promedio es de {peor_tienda_data['Calificacion_Prom']:.2f}/5.0.")
print(f"- Eficiencia Logística: El costo promedio de envío es de ${peor_tienda_data['Envio_Prom']:,.2f}.")

print(f"\n3. ANÁLISIS DE LA TIENDA DE BAJO RENDIMIENTO ({peor_tienda_nombre})")
print("Al observar el gráfico de pastel, vemos la distribución de sus categorías. ")
print("El gráfico de dispersión revela la falta de productos 'Estrella' (alto precio y alta venta).")

print(f"\n4. CONCLUSIÓN Y RECOMENDACIÓN")
print("Basado en los datos, se recomienda al Sr. Juan VENDER la **" + peor_tienda_nombre + "**.")
print("Esta tienda no solo genera el menor flujo de caja, sino que sus métricas operativas")
print("sugieren dificultades para escalar en comparación con las otras opciones.")
print("="*80)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from folium.plugins import HeatMap

# 1. PREPARACIÓN DE DATOS (Etiquetar y Unir)
# Se añade columna ID a cada dataframe
tienda['ID_Tienda'] = 'Tienda 1'
tienda2['ID_Tienda'] = 'Tienda 2'
tienda3['ID_Tienda'] = 'Tienda 3'
tienda4['ID_Tienda'] = 'Tienda 4'

df_total = pd.concat([tienda, tienda2, tienda3, tienda4])

# 2. VISUALIZACIÓN ESTÁTICA
plt.figure(figsize=(10, 8))
# Usamos seaborn porque maneja mejor los colores por categoría automáticamente
sns.scatterplot(
    data=df_total,
    x='lon',
    y='lat',
    hue='ID_Tienda',
    alpha=0.5,
    palette='bright'
)

plt.title('Distribución Geográfica de Ventas por Tienda')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.legend(title='Tienda')
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()

# 3. VISUALIZACIÓN INTERACTIVA
# Centrar el mapa en el promedio de las coordenadas
mapa_calor = folium.Map(location=[df_total['lat'].mean(), df_total['lon'].mean()], zoom_start=5)

# Creación de una capa de calor basada en la densidad de ventas
heat_data = [[row['lat'], row['lon']] for index, row in df_total.iterrows()]

HeatMap(heat_data, radius=10, blur=15).add_to(mapa_calor)

print("\n--- Mapa de Calor Interactivo generado (abajo) ---")
mapa_calor
```

La visualización espacial revela una fuerte polarización de la actividad comercial. Se observa una densa concentración de ventas en los principales clústeres urbanos (Bogotá, Medellín, Cali y la Costa Caribe), lo que favorece a las tiendas posicionadas en estas zonas mediante una logística eficiente y costos de envío reducidos.

Por el contrario, la dispersión de ventas hacia zonas periféricas y de difícil acceso (como la región amazónica o los llanos orientales) impone una carga operativa significativa: los costos de envío se disparan y los tiempos de entrega se alargan. Esto crea una correlación inversa entre la distancia al centro urbano y la rentabilidad del pedido, afectando negativamente tanto al margen de ganancia como a la calificación de satisfacción del cliente debido a las fricciones logísticas.
